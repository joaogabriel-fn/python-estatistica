## **Importando e Analisando a Base do MovieLens**

### **Passo 1** - Importando os dados

```python
notas = pd.read_csv('ratings.csv')
notas.head() 
```

### **Passo 2** - Agrupar os dados de acordo com o Id do filme, a fim de obter a nota média

```python
nota_media_por_filme = notas.groupby('movieId')[['rating']].mean()
nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme['rating'], 
                  stat='density', 
                  kde=True, 
                  lw=0, 
                  kde_kws=dict(cut=3)
)

ax.set(xlabel='Nota Media', ylabel='Densidade')
ax.set_title('Media de votos em filmes no Movielens 100k');
```

### **Passo 3** - Filtrar os filmes com pelo menos 10 votos

```python
qtd_votos_por_filme = notas.groupby('movieId').count()
filmes_com_10_ou_mais_votos = qtd_votos_por_filme.query('rating >= 10').index.values
filmes_com_10_ou_mais_votos
```

```python
nota_media_de_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_10_ou_mais_votos]
```

```python
ax = sns.histplot(nota_media_de_filmes_com_pelo_menos_10_votos, 
                  stat='density', 
                  kde=True, 
                  lw=0, 
                  kde_kws=dict(cut=3)
)

ax.set(xlabel='Nota Media', ylabel='Densidade')
ax.set_title('Media de votos em filmes no Movielens 100k com 10 ou mais votos');
```

```python
ax = sns.boxplot(nota_media_de_filmes_com_pelo_menos_10_votos, orient='h')
ax.set(xlabel='Nota Média do Filme')
ax.set_title('Distribuição de nota média dos filmes do Movielens com 10 ou mais votos');
```

```python
ax = sns.histplot(nota_media_de_filmes_com_pelo_menos_10_votos, 
                  stat='density', 
                  cumulative=True,
                  kde=True, 
                  lw=0, 
                  kde_kws=dict(cut=3)
)

ax.set(xlabel='Nota Media', ylabel='Densidade')
ax.set_title('Media de votos em filmes no Movielens 100k com 10 ou mais votos');
```

### **Passo 4** - Analisar os dados das amostras filtradas

```python
print(f'Media dos filmes com pelo menos 10 votos: {nota_media_de_filmes_com_pelo_menos_10_votos.mean().mean():.2f}')
```

```python
import matplotlib.pyplot as plt

temp = nota_media_de_filmes_com_pelo_menos_10_votos.sample(frac=1, random_state=75243)

medias = [temp[0:i].mean().mean() for i in range(1, len(temp))]

plt.plot(medias);
```

```python
from statsmodels.stats.weightstats import zconfint, DescrStatsW

zconfint(nota_media_de_filmes_com_pelo_menos_10_votos)
```

```python
descr_todos_com_10_votos = DescrStatsW(nota_media_de_filmes_com_pelo_menos_10_votos)
```

```python
descr_todos_com_10_votos.tconfint_mean()
```

### **Passo 1** - Analisando a média de notas de 1 filme

```python
filmes = pd.read_csv('movies.csv')
filmes.head()
```

```python
filmes.query('movieId==1')
```

```python
notas1 = notas.query('movieId == 1')
notas1.head() 
```

```python
ax = sns.histplot(notas1['rating'], 
                  stat='density', 
                  kde=True, 
                  lw=0, 
                  kde_kws=dict(cut=3)
)

ax.set(xlabel='Nota Media', ylabel='Densidade')
ax.set_title('Nota para o Toy Story');
```

```python
ax = sns.boxplot(notas1['rating'], orient='h')
ax.set(xlabel='Nota')
ax.set_title('Distribuição de notas para Toy Story');
```

```python
notas1['rating'].mean()
```

```python
notas1['rating'].count()
```

```python
zconfint(notas1['rating'])
```

```python
from statsmodels.stats.weightstats import ztest

z, p_valor = ztest(notas1['rating'], value = nota_media_de_filmes_com_pelo_menos_10_votos.mean().mean())
```

```python
if p_valor <= 0.05:
    print('Rejeitar H0, as médias não são iguais')
else:
    print('Aceitar H0')
```

Conclusão: Ao analisar o filme Toy Story, é possível aplicar um z-teste para afirmar que sua média é diferente da média geral dos filmes

```python
temp = notas1.sample(frac=1, random_state=75241)['rating']

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias);
```

```python
import numpy as np

temp = notas1.sample(frac=1, random_state=75241)['rating']

def calcula_teste(i):
    media = temp[0:i].mean()
    _, p_valor = ztest(temp[0:i], value = nota_media_de_filmes_com_pelo_menos_10_votos.mean().mean())
    return (i, media, p_valor)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1]);
plt.plot(valores[:,0], valores[:,2]);
plt.hlines(y=0.05, xmin=2, xmax=len(temp), colors='r');
```

```python
print(ztest(notas1['rating'], notas['rating']))
zconfint(notas1['rating'], notas['rating'])
```

```python
from scipy.stats import ttest_ind

ttest_ind(notas1['rating'], notas['rating'])
```
